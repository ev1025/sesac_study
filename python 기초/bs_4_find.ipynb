```python
# 강사님 답
import requests
from bs4 import BeautifulSoup
# URL 요청 및 BeautifulSoup 객체 생성
url = 'https://finance.naver.com/sise/sise_quant.nhn'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
# 상승, 하락, 보합 여부를 포함한 데이터 수집
def extract_stock_data(tag):
    """하나의 tr 태그에서 데이터를 추출"""
    tds = tag.select('td')
    status = '변화 없음'
    if tds[3].find('span') and tds[3].find('span').text.strip() != '보합':
        status = tds[3].select_one('span').text.strip()
    return {
        "순위": tds[0].text.strip(),
        "종목": tds[1].text.strip(),
        "현재가": tds[2].text.strip(),
        "전일비": tds[3].select('span')[1].text.strip() if len(tds[3].select('span')) > 1 else '',
        "등락여부": status
    }
# 유효한 tr 태그를 필터링하여 데이터 수집
stock = [
    extract_stock_data(tag)
    for tag in soup.select('table.type_2 tr')
    if len(tag.select('td')) == 12
]
# 결과 출력
print(f"총 데이터 수: {len(stock)}")
print(stock[:5])  # 일부 데이터 확인
```

```python
# 전체 데이터 가져오기 (원본)
# tqdm 사용하기
import pandas as pd
import requests
from bs4 import BeautifulSoup
from datetime import datetime

# 오늘 날짜
today = datetime.now()
formatted_date = today.strftime('%Y%m%d')


# 전체 페이지 번호를 받기 위해 값을 가져옴
response = requests.get(f'https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid=032&date={formatted_date}&page=1')
soup = BeautifulSoup(response.text, 'html.parser')

# 페이지 데이터
page_data = soup.select('div.paging')

# 마지막 페이지
last_num = int(page_data[0].text.strip().split('\n')[-1])

# 값 받을 리스트
all_list =  []


# 페이지 돌리기
for page_num in range(1, last_num+1):
    url = f'https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid=032&date={formatted_date}&page={page_num}'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    all_data = soup.select('div#main_content')[0]

    # 본문 데이터
    head_data = all_data.select('.list_body.newsflash_body')
    
    # 현재 페이지 번호
    current_num = int(page_data[0].select_one('strong').text)


    if page_num <= last_num:
        # 사이트 주소 리스트 가져오기
        url_list = []
        url_list.extend(head_data[0].select('dt a.nclicks\(cnt_papaerart\)'))
        url_list.extend(head_data[0].select('dt a.nclicks\(cnt_flashart\)'))
        
        # 사이트 주소 리스트 만들기
        site_list = []
        for site in url_list:
            if site.get('href') not in site_list:
                site_list.append(site.get('href'))

        # 사이트 리스트 추가(set으로 중복 사이트명 제거)
        for j in site_list:
            response = requests.get(j)
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # 본문 전체 가져오기
            contents_text = soup.select('div#ct')[0]
            all_list.append({"page" : page_num,
                            "title": contents_text.select('#title_area')[0].text,
                            "body" : contents_text.select('#dic_area')[0].text.strip().replace("\n","")})


final_list = pd.DataFrame(all_list)
final_list.to_csv('final_list.csv', index=False)
```

```python
# 전체 데이터 가져오기
import pandas as pd
import requests
from bs4 import BeautifulSoup
from datetime import datetime

# 오늘 날짜
today = datetime.now()
formatted_date = today.strftime('%Y%m%d')

# 첫 페이지 번호
page_ = 1

# 전체 페이지 번호를 받기 위해 값을 가져옴
response = requests.get(f'https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid=032&date=20241223&page={page_}')
soup = BeautifulSoup(response.text, 'html.parser')

# 페이지 데이터
page_data = soup.select('div.paging')

# 마지막 페이지 찾기
while page_data[0].text.strip().split('\n')[-1] == '다음':
    page_ +=10
    response = requests.get(f'https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid=032&date=20241223&page={page_}')
    soup = BeautifulSoup(response.text, 'html.parser')
    page_data = soup.select('div.paging')

last_num = int(page_data[0].text.strip().split('\n')[-1])

# 값 받을 리스트
all_list =  []

# 수집 실패한 url 리스트 
fail_list = []

# 페이지 돌리기
for page_num in range(1, last_num+1):
    url = f'https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid=032&date=20241223&page={page_num}'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    all_data = soup.select('div#main_content')[0]

    # 본문 데이터
    head_data = all_data.select('.list_body.newsflash_body')
    
    # 현재 페이지 번호
    current_num = int(page_data[0].select_one('strong').text)


    if page_num <= last_num:
        # 사이트 주소 리스트 가져오기
        url_list = []
        url_list.extend(head_data[0].select('dt a.nclicks\(cnt_papaerart\)'))
        url_list.extend(head_data[0].select('dt a.nclicks\(cnt_flashart\)'))
        
        
        # 사이트 주소 리스트 만들기
        site_list = []
        for site in url_list:
            if site.get('href') not in site_list:
                site_list.append(site.get('href'))

        # 사이트 리스트 추가(set으로 중복 사이트명 제거)
        for j in site_list:
            response = requests.get(j)
            soup = BeautifulSoup(response.text, 'html.parser')

            # 본문 전체 가져오기
            try:
                contents_text = soup.select('div#ct')[0]
                all_list.append({"page" : page_num,
                                    "title": contents_text.select('#title_area')[0].text,
                                    "body" : contents_text.select('#dic_area')[0].text.strip().replace("\n","")})
            # 안 긁힌 경우 fail_list에 사이트 주소 넣기
            except:
                fail_list.append(j)
                continue


final_list = pd.DataFrame(all_list)
final_list.to_csv('final_list.csv', index=False)
final_list.tail()
```

```python
# 강사님 크롤링


from tqdm import tqdm
import requests
from bs4 import BeautifulSoup
# 기본 URL 설정
BASE_URL = 'https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid=032&page={}'
def get_news_links(page):
    """
    지정된 페이지에서 뉴스 링크 수집
    """
    response = requests.get(BASE_URL.format(page))
    soup = BeautifulSoup(response.text, 'html.parser')
    # type06으로 시작하는 클래스의 뉴스 링크 추출
    li_tags = soup.select('div.list_body.newsflash_body ul[class^="type06"] li')
    links = [li_tag.select_one('a').attrs['href'] for li_tag in li_tags if li_tag.select_one('a')]
    return links
def scrape_news_content(url):
    """
    뉴스 링크에서 제목과 본문 크롤링
    """
    news_res = requests.get(url)
    news_soup = BeautifulSoup(news_res.text, 'html.parser')
    # 뉴스 제목과 본문 추출
    title_element = news_soup.select_one('#title_area span')
    body_element = news_soup.find(id='dic_area')
    if title_element and body_element:
        title = title_element.text.strip()
        body = ''.join(body_element.find_all(string=True, recursive=False)).strip()
        return {"title": title, "body": body, "link": url}
    else: return None
def crawl_all_news():
    """
    링크를 먼저 수집하고, 뉴스를 크롤링
    """
    all_links = []
    page = 1
    last_page_links = None  # 이전 페이지 링크 데이터를 저장
    # 페이지 크롤링 tqdm 설정
    with tqdm(desc="Fetching links", unit="page") as page_pbar:
        while True:
            page_pbar.set_description(f"Fetching links from page {page}")
            current_links = get_news_links(page)
            # 이전 페이지와 동일한 링크면 중단
            if not current_links or current_links[-1] == last_page_links:
                print(f"No more new links found or duplicate links on page {page}. Stopping.")
                break
            # 새로운 링크 추가
            all_links.extend(current_links)
            last_page_links = current_links[-1]
            page += 1
            page_pbar.update(1)
    # 뉴스 크롤링 tqdm 설정
    print(f"Total links collected: {len(all_links)}")
    all_news = []
    failed_links = []  # 크롤링 실패한 링크를 저장할 리스트
    with tqdm(total=len(all_links), desc="Scraping news", unit="article") as news_pbar:
        for link in all_links:
            news_content = scrape_news_content(link)
            if news_content:
                all_news.append(news_content)
            else:
                failed_links.append(link)  # 실패한 링크 저장
            news_pbar.update(1)
    return all_news, failed_links
if __name__ == "__main__":
    # 크롤링 실행
    all_news, failed_news = crawl_all_news()
    # 결과 출력
    print(f"총 뉴스 기사 수: {len(all_news)}")
    print(all_news[:5])  # 첫 5개의 뉴스 출력
```

```python
import requests
from bs4 import BeautifulSoup
import jwt
import datetime
def decode_token(token):
    """JWT 토큰 디코딩 및 만료 시간 반환"""
    try:
        decoded = jwt.decode(token, options={"verify_signature": False})
        exp = datetime.datetime.fromtimestamp(decoded.get('exp'))
        print(f"토큰 만료 시간: {exp}")
        if exp < datetime.datetime.now():
            print(":경고: 토큰이 만료되었습니다. 새 토큰을 발급받으세요.")
        return exp
    except Exception as e:
        print(f"토큰 디코딩 중 오류 발생: {e}")
        return None
def fetch_article(url):
    """기사 제목과 본문 가져오기"""
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        title = soup.select_one('h3.tit_view').text.strip()
        body = ' '.join(p_tag.text.strip() for p_tag in soup.select('div.article_view section p'))
        return title, body
    except Exception as e:
        print(f"기사 데이터 가져오기 중 오류 발생: {e}")
        return None, None
def fetch_reactions(item_key, token):
    """리액션 데이터 가져오기"""
    try:
        headers = {'Authorization': f'Bearer {token}'}
        response = requests.get(f'https://action.daum.net/apis/v1/reactions/home?itemKey={item_key}', headers=headers)
        response.raise_for_status()
        return response.json().get('item', {}).get('stats', {})
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP 오류 발생: {http_err}")
    except Exception as e:
        print(f"리액션 데이터를 가져오는 중 오류 발생: {e}")
    return {}
# 설정
ARTICLE_URL = 'https://v.daum.net/v/20240811221440047'
ITEM_KEY = '20240811221440047'
BEARER_TOKEN = 'eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJqdGkiOiJjOGM3YWM1My1mMTJhLTQyNTEtOWYyNS1mODEyMTc5YjgxYzYiLCJjbGllbnRfaWQiOiIyNkJYQXZLbnk1V0Y1WjA5bHI1azc3WTgiLCJmb3J1bV9rZXkiOiJuZXdzIiwiZm9ydW1faWQiOi05OSwiZ3JhbnRfdHlwZSI6ImFsZXhfY3JlZGVudGlhbHMiLCJhdXRob3JpdGllcyI6WyJST0xFX0NMSUVOVCJdLCJzY29wZSI6W10sImV4cCI6MTczNTA0ODY4N30.izCiGv9f6riFvuXGruUrB2toME5doS23BqECK34-zO4'
# 토큰 만료 확인
decode_token(BEARER_TOKEN)
# 기사 데이터 가져오기
title, body = fetch_article(ARTICLE_URL)
if title and body:
    print(f"제목: {title}\n본문: {body[:100]}... (생략)\n")
# 리액션 데이터 가져오기
reactions = fetch_reactions(ITEM_KEY, BEARER_TOKEN)
# 결과 정리
result = {
    "title": title or "제목 없음",
    "body": body or "본문 없음",
    "reactions": {react: reactions.get(react, 0) for react in ['RECOMMEND', 'LIKE', 'IMPRESS', 'ANGRY', 'SAD']}
}
# 결과 출력
print(result)
```
